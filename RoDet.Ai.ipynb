```python
import cv2
from ultralytics import YOLO
model = YOLO(r"C:\Users\BIT\mouse detector\runs\detect\train2\weights\best.pt")
logo = cv2.imread(r"C:\Users\BIT\Downloads\logo-removebg-preview (1).png", cv2.IMREAD_UNCHANGED)  # Make sure logo.png exists
logo = cv2.resize(logo, (100, 100))  
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)
    annotated_frame = results[0].plot()
    x_offset, y_offset = 10, 10
    y1, y2 = y_offset, y_offset + logo.shape[1]
    x1, x2 = x_offset, x_offset + logo.shape[1]
    if logo.shape[2] == 4:
        alpha_logo = logo[:, :, 3] / 255.0
        for c in range(3):
            annotated_frame[y1:y2, x1:x2, c] = (
                alpha_logo * logo[:, :, c] + 
                (1 - alpha_logo) * annotated_frame[y1:y2, x1:x2, c]
            )
    else:
        annotated_frame[y1:y2, x1:x2] = logo
    cv2.imshow("RoDet.Ai - Live", annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
